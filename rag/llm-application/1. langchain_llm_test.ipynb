```python
from dotenv import load_dotenv

load_dotenv()
```

```python
from langchain_openai import ChatOpenAI



llm = ChatOpenAI(
    model="gpt-3.5-turbo",      # 또는 "gpt-4o", "gpt-3.5-turbo"
    temperature=0.7,
    max_retries=3,            # 요청 실패 시 3번 재시도
    request_timeout=60        # 최대 60초 대기
)
```

```python
ai_message = llm.invoke("인프런에 어떤 강의가 있나요?")
```
