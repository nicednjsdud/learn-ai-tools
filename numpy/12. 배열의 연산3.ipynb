### 3-3. 배열의 연산3 : min/max/sum/mean/std/cumsum/median

```python
import numpy as np
```

```python
arr = np.array([[1, 2, 3], 
                [4, 5, 6]])
```

```python
# min()

print(np.min(arr))
print(arr.min())
print(arr.min(axis=0)) # Column-wise minimum
```

```python
# max()

print(np.max(arr))
print(arr.max())
print(arr.max(axis=0)) # Row-wise maximum
```

```python
# sum()

print(np.sum(arr))
print(arr.sum())
print(arr.sum(axis=0)) # Row-wise sum
```

```python
# mean()
# 평균값

print(np.mean(arr))
print(arr.mean())
print(arr.mean(axis=0)) # Column-wise mean
```

```python
# std()
# 표준편차

print(np.std(arr))
print(arr.std())
print(arr.std(axis=0)) # Column-wise standard deviation
```

```python
# cumsum()
# 누적합

print(np.cumsum(arr))
print(arr.cumsum())
print(arr.cumsum(axis=0)) # Column-wise cumulative sum
```

```python
# median()
# 중앙값

print(np.median(arr))
print(arr)
print(np.median(arr, axis=0)) # Column-wise median
```
