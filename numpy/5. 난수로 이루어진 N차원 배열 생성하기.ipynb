```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# np.random.normal()

# arr = np.random.normal(loc, scale, size)

arr = np.random.normal(0, 1, (2, 3))
print(arr)
```

```python
# plt example

arr = np.random.normal(0 , 1 , 100000)
plt.hist(arr, bins=100)
plt.show()
```

```python
# np.random.rand()
# 0~1 uniform distribution (균등 분포)
arr = np.random.rand(1000)
plt.hist(arr, bins=100)
plt.show()
```

```python
# np.random.randn()
# normal distribution (정규 분포)
arr = np.random.randn(1000)
plt.hist(arr, bins=100)
plt.show()
```

```python
# np.random.randint()
# random integers from low (inclusive) to high (exclusive)
# np.random.randint(low, high=None, size=None, dtype=int)

arr = np.random.randint(low= 1 , high = 5, size=(3, 4))
print(arr, end = "\n\n")

arr = np.random.randint(10)
print(arr)
```

```python
# 100 ~ 200 사이의 정수 1000개
arr = np.random.randint(100, 200, 1000)
plt.hist(arr, bins=100)
plt.show()
```
