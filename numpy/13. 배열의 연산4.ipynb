### 3-4. 배열의 연산4 : 비교 연산/삼각 함수

```python
import numpy as np
```

```python
# 비교 연산
arr1 = np.array([[1, 2, 3], 
                 [4, 5, 6]])
arr2 = np.array([[1, 0, 3], 
                 [4, -2, 9]])

# 두 배열의 요소별 비교
print(arr1 == arr2, end='\n\n')

# 두 배열의 대소 비교
print(arr1 > arr2, end='\n\n')

# equal 함수 사용
print(np.array_equal(arr1, arr2))
```

```python
# 삼각 함수
arr = np.array([[1, 2, 3],
                [4, 5, 6]])

# 사인 함수
print(np.sin(arr), end='\n\n')

# 코사인 함수
print(np.cos(arr), end='\n\n')

# 탄젠트 함수
print(np.tan(arr), end='\n\n')

# pi 상수
print(np.pi)
```
