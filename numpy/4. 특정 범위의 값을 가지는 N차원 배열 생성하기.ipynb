### 1-4. 특정 범위의 값을 가지는 N차원 배열 생성하기

```python
import numpy as np

lst = list(range(0 , 9 , 2))
print(lst)
```

```python
# Using numpy arange() function
arr = np.arange(0 , 9 , 2)
print(arr)
arr = np.arange(3, 12)
print(arr)
arr = np.arange(stop= 9, start= 0 , step= 2) # keyword arguments
print(arr)
```

```python
# Using numpy linspace() function
# np.linspace(start, stop, num=50)
arr = np.linspace(0 , 100 ,120)
print(arr)
```

```python
# Using numpy logspace() function
# np.logspace(start, stop, num=50, base=10.0)

arr = np.linspace(1, 10, 10)
print(arr, end = "\n\n")

arr = np.logspace(1, 10, 10, base=2)
print(arr)

arr = np.logspace(0, 5, 6) # default base is 10
print(arr)
```
