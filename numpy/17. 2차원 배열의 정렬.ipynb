#### 4-2. 2차원 배열의 정렬

```python
import numpy as np
```

```python
arr = np.random.randint(15, size=(4, 4))
print(arr)
```

```python
print(np.sort(arr, axis=0))  # Sort each column
```

```python
print(np.sort(arr, axis=1))  # Sort each row
```

```python
print(np.sort(arr, axis=None))  # Sort the flattened array
```

```python
# argsort()
# argsort는 정렬된 배열의 인덱스를 반환합니다.
print(np.sort(arr, axis=1), end="\n\n")
print(np.argsort(arr, axis=1))
```

```python
print (np.sort(arr, axis=0), end="\n\n")
print (np.argsort(arr, axis=0))
```
