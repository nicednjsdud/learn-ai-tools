### 1-6. 시드(Seed)값을 통한 난수 생성 제어

```python
import numpy as np
```

```python
arr = np.random.rand(10)
print("난수 발생1 \n", arr)

arr = np.random.rand(10)
print("난수 발생2 \n", arr)
```

```python
# 시드값 설정
# 같은 시드값을 주면 같은 난수 발생
# 시드값을 주지 않으면 매번 다른 난수 발생
np.random.seed(1)
arr = np.random.rand(10)
print("난수 발생1 \n", arr)

np.random.seed(1)
arr = np.random.rand(10)
print("난수 발생2 \n", arr)
```
