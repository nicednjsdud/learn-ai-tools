#### 6-3. 배열 분합

```python
import numpy as np
```

```python
# 2차원 배열
arr = np.arange(1, 13).reshape(3, 4)
print("Original Array:")
print(arr)

# vsplit: 수직으로 배열을 나눔
vsplit_arr = np.vsplit(arr, 3)
print("\nVsplit Result:")
for sub_arr in vsplit_arr:
    print(sub_arr)

# hsplit: 수평으로 배열을 나눔
hsplit_arr = np.hsplit(arr, 2)
print("\nHsplit Result:")
for sub_arr in hsplit_arr:
    print(sub_arr)
```

```python
# 3차원 배열
arr_3d = np.random.randint(0, 10, (4, 6, 8))
print(arr_3d)

```

```python
arr_vsplit_3d = np.vsplit(arr_3d, 2)
print("\n3D Array Vsplit Result:")
for sub_arr in arr_vsplit_3d:
    print(sub_arr)
```

```python
hsplit_arr_3d = np.hsplit(arr_3d, 3)
print("\n3D Array Hsplit Result:")
for sub_arr in hsplit_arr_3d:
    print(sub_arr)
```
