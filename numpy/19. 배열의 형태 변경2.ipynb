#### 5-1. 배열의 형태 변경2: resize(), revel()

```python
import numpy as np
```

```python
arr = np.arange(12)

print(arr)
```

```python
# resize()

arr.resize((3, 4))
print("\nResized array (3x4):\n", arr)
```

```python
# ravel()

raveled_arr = arr.ravel()
print("\nRaveled array:\n", raveled_arr)
```

```python

# reshape() -1

arr = np.arange(1, 13)
print("\nOriginal array:\n", arr)

# 자동으로 열의 크기를 계산하여 3행으로 재구성
reshaped_arr = arr.reshape(3, -1)
print("\nReshaped array (3 rows, inferred columns):\n", reshaped_arr)

# 3차원 배열로 재구성
reshaped_arr_3d = arr.reshape(3, 2, -1)
print("\nReshaped array (3x2x2):\n", reshaped_arr_3d)
```
