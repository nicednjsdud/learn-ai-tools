### 3-6. 백터 연산의 장점

* 벡터 연산의 장점은 한 줄로 말하면 **“빠르고, 간단하고, 안정적이다”**.
* NumPy·PyTorch·TensorFlow가 다 벡터화(Vectorization)를 강조하는 이유가 바로 이것들 때문.

#### ✔ 벡터 연산의 4가지 핵심 장점

1) 속도가 압도적으로 빠르다 (성능)
2) 코드가 간결해지고 가독성이 좋아짐
3) 버그 발생 가능성이 줄어든다
4) GPU 가속을 이용하기 쉬움 (딥러닝 필수)

```python
import numpy as np
```

```python
import time
```

```python
arr = np.arange(99999999)
```

```python
# for
sum = 0
before = time.time()
for i in arr:
    sum += i

after = time.time()
print("for:", after - before, "seconds")
```

```python
# 벡터 연산
before = time.time()
sum = np.sum(arr)
after = time.time()
print("vectorized:", after - before, "seconds")
```

```python
# 배열 
arr1 = np.arange(99999999)
arr2 = np.arange(99999999)
```

```python
# for
sum = 0
before = time.time()
for i, j in zip(arr1, arr2):
    sum += i * j
after = time.time()
print("for (multiple):", after - before, "seconds")
```

```python
# 벡터 연산
before = time.time()
sum = np.sum(arr1 * arr2)
after = time.time()
print("vectorized (multiple):", after - before, "seconds")
```
