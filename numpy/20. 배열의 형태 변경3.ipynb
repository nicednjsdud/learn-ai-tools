#### 5-3. 배열의 형태3 : expand_dims(), squeeze()

#### expand_dims() — 차원을 “추가”하는 함수

* 특정 위치(axis)에 새로운 차원(길이 1)을 추가하는 함수.

#### squeeze() — 길이 1 차원을 “제거”하는 함수

* shape에서 길이가 1인 차원만 제거하는 함수.

```python
import numpy as np
```

```python
# expand_dims()

arr = np.array([1, 2])
print("Original array:")
print(arr, arr.shape)

arr = np.expand_dims(arr, axis=0)
print("\nArray after expanding dimensions at axis 0:")
print(arr, arr.shape)

arr = np.expand_dims(arr, axis=1)
print("\nArray after expanding dimensions at axis 1:")
print(arr, arr.shape)
```

```python
# squeeze()

arr = np.array([[1, 2]])
print("\nOriginal array:")
print(arr, arr.shape, arr.ndim)

arr = np.squeeze(arr, axis=0)
print("\nArray after squeezing:")
print(arr, arr.shape, arr.ndim)

arr = np.array([[[1], 
                 [2], 
                 [3]]])

print("\nOriginal array:")
print(arr, arr.shape, arr.ndim)

arr = np.squeeze(arr, axis=0)
print("\nArray after squeezing:")
print(arr, arr.shape, arr.ndim)

arr = np.array([[[1, 2, 3]]])
print("\nOriginal array:")
print(arr, arr.shape, arr.ndim)

arr = np.squeeze(arr, axis=1)
print("\nArray after squeezing:")
print(arr, arr.shape, arr.ndim)
```
