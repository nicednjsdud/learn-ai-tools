#### 6-2 배열 간의 병합 

```python
import numpy as np
```

```python
# append()

arr1 = np.arange(1, 13).reshape(3, 4)
arr2 = np.arange(13, 25).reshape(3, 4)

print("Array 1:")
print(arr1)
print("\nArray 2:")
print(arr2)
```

```python
# axis=0 (vertical append)
appended_axis0 = np.append(arr1, arr2, axis=0)
print("\nAppended along axis 0:")
print(appended_axis0)
```

```python
# axis=1 (horizontal append)
appended_axis1 = np.append(arr1, arr2, axis=1)
print("\nAppended along axis 1:")
print(appended_axis1)
```

```python
# No axis (flattened append)
appended_no_axis = np.append(arr1, arr2)
print("\nAppended with no axis (flattened):")
print(appended_no_axis)
```

```python
# vstack(), hstack()
arr1 = np.arange(1, 7).reshape(2, 3)
arr2 = np.arange(7, 13).reshape(2, 3)

arr3 = np.vstack((arr1, arr2))
print("\nVertical Stack (vstack):")
print(arr3)

arr4 = np.hstack((arr1, arr2))
print("\nHorizontal Stack (hstack):")
print(arr4)
```

```python
# concatenate()
arr1 = np.arange(1, 10).reshape(3, 3)
arr2 = np.arange(10, 19).reshape(3, 3)

concatenated_axis0 = np.concatenate((arr1, arr2), axis=0)
print("\nConcatenated along axis 0:")
print(concatenated_axis0)

concatenated_axis1 = np.concatenate((arr1, arr2), axis=1)
print("\nConcatenated along axis 1:")
print(concatenated_axis1)
```
